```python
#importing all the necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#This reads a CSV file containing Olympic history and athlete results from a Kaggle dataset and stores it in the DataFrame df for further analysis.

df = pd.read_csv('/kaggle/input/120-years-of-olympic-history-athletes-and-results/athlete_events.csv')
```

```python
#The expression retrieves the count of non-missing (non-NaN) values for each column in the DataFrame df

df.describe(include='all').loc['count']
```

```python
# The expression returns the total number of missing values (NaN) for each column in the DataFrame 

df.isnull().sum()
```

```python
#The expression calculates the percentage of missing values for each column in the DataFrame df. 

df.isnull().mean()*100
```

```python
# Importing Missing Number Pacakage as msno

import missingno as msno
```

# 1. Bar Plot

The bar plot offers a side-by-side comparison of missing values across different columns, making it useful for quickly identifying which columns might need imputation or removal based on missingness.

```python
msno.bar(df)
```

**Observations:**

There are some values missing in column of Age and approximately about 25% missing values in columns of Height and weight and more than 80%-90% values missing in column of Medal which is also consistent with our statistical findings

# 2. Matrix Plot

The matrix plot in the missingno package is a visualization tool used to detect and display missing data in a dataset.

```python
msno.matrix(df)
```

**Observations:**

The white lines in certain columns (e.g., Age, Height, Weight, etc.) indicate missing data, while black areas indicate non-missing data.
Columns like Medal and Age seem to have the most missing values, while columns such as ID, Name, NOC, and Sport appear to have very few or no missing values.

# 3. Heatmap
The heatmap is used to identify correlations of the nullity between each of the different columns. In other words, it can be used to identify if
there is a relationship in the presence of null values between each of the columns.

```python
msno.heatmap(df)
```

**Observations:**
* High correlation (0.9) between missing values in Height and Weight, indicating that when data is missing for one, its likely missing for the other as well.
* There is some correlation (~0.3) between missing values in Age and Height/Weight, though it is not as strong.
* No missing correlation for the 'Medal' column, as the corresponding area is blank (white), indicating that 'Medal' likely has no missing values or no correlation with missing values in other columns.
* We can see that there is no correlation between the missingness in Age and Medal with rest of the variables.


# 4. Dendogram

The dendrogram plot provides a tree-like graph generated through hierarchical clustering and groups together columns that have strong correlations in nullity.


```python
msno.dendrogram(df)
```

**Observations:**

* The columns such as Age, Weight, and Height seem to have a significant clustering of missing values, as indicated by their proximity in the dendrogram.
Other columns like Sport, Event, and Medal are clustered together, suggesting they may also share similar missing data patterns.
Columns such as Name, ID, and Sex seem to have fewer missing values, given their isolated position in the dendrogram.
